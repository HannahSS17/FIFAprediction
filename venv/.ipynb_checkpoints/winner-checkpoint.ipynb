```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython import get_ipython
get_ipython().run_line_magic('matplotlib', 'inline')

# % matplotlib inline


df = pd.read_excel("FullData.xls", error_bad_lines=False)
df.head(7)

# delete unnecessary columns from data file
del df['National Kit']
df.head()

plt.figure(figsize=(15,32))

#plot all nations

#set nationalities for y exis
sns.countplot(y = df.Nationality, palette="Set2")

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython import get_ipython
get_ipython().run_line_magic('matplotlib', 'inline')

# % matplotlib inline


df = pd.read_excel("FullData.xls", error_bad_lines=False)
df.head(7)

# delete unnecessary columns from data file
del df['National Kit']
df.head()

plt.figure(figsize=(15,32))

#plot all nations

#set nationalities for y exis
sns.countplot(y = df.Nationality, palette="Set2")

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython import get_ipython
get_ipython().run_line_magic('matplotlib', 'inline')

# % matplotlib inline


df = pd.read_excel("FullData.xls", error_bad_lines=False)
df.head(7)

# delete unnecessary columns from data file
del df['National Kit']
df.head()

plt.figure(figsize=(15,32))

#plot all nations

#set nationalities for y exis
sns.countplot(y = df.Nationality, palette="Set2")

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython import get_ipython
get_ipython().run_line_magic('matplotlib', 'inline')

# % matplotlib inline


df = pd.read_excel("FullData.xls", error_bad_lines=False)
df.head(7)

# delete unnecessary columns from data file
del df['National Kit']
df.head()

plt.figure(figsize=(15,32))

#plot all nations

#set nationalities for y exis
sns.countplot(y = df.Nationality, palette="Set2")

```
